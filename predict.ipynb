Carregamento das Ferramentas e do Modelo

```python
import pandas as pd
import joblib

# Carregando o modelo que você salvou no estágio de treinamento
model = joblib.load('models/model.pkl')

print("Modelo carregado com sucesso!")
```

A Função de Predição Inteligente

```python
def prever_atraso(dados_pedido):
    """
    Recebe os dados de um novo pedido e retorna a probabilidade de atraso.
    """
    # Criando o DataFrame para o modelo
    df_novo = pd.DataFrame([dados_pedido])
    
    # Realizando a predição
    probabilidade = model.predict_proba(df_novo)[0][1]
    previsao = model.predict(df_novo)[0]
    
    resultado = "ATRASO PROVÁVEL" if previsao == 1 else "NO PRAZO"
    
    print(f"--- Resultado da Análise ---")
    print(f"Status Previsto: {resultado}")
    print(f"Risco de Atraso: {probabilidade:.2%}")

print("Função de predição pronta para uso.")
```

Simulação de um Pedido de Alto Risco

```python
# Simulação: Pedido pesado, interestadual e com prazo curto
meu_pedido = {
    'price': 350.00,
    'freight_value': 85.50,
    'product_weight_g': 15000.0,    # 15kg
    'product_volume_cm3': 60000.0,  # Grande
    'is_interstate': 1,             # Entre estados diferentes
    'estimated_days_to_deliver': 5, # Prazo apertado
    'purchase_day_of_week': 4       # Sexta-feira
}

prever_atraso(meu_pedido)
```
