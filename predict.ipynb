Carregamento das Ferramentas e do Modelo

```python
import pandas as pd
import joblib

# Carregando o modelo que você salvou no estágio de treinamento
model = joblib.load('models/model.pkl')

print("Modelo carregado com sucesso!")
```

A Função de Predição Inteligente

```python
def prever_atraso(dados_pedido):
    """
    Recebe os dados de um novo pedido e retorna a probabilidade de atraso.
    """
    # Criando o DataFrame para o modelo
    df_novo = pd.DataFrame([dados_pedido])
    
    # Realizando a predição
    probabilidade = model.predict_proba(df_novo)[0][1]
    previsao = model.predict(df_novo)[0]
    
    resultado = "ATRASO PROVÁVEL" if previsao == 1 else "NO PRAZO"
    
    print(f"--- Resultado da Análise ---")
    print(f"Status Previsto: {resultado}")
    print(f"Risco de Atraso: {probabilidade:.2%}")

print("Função de predição pronta para uso.")
```

Simulação de um Pedido de Alto Risco

```python
# Pedido pesado, caro, cruzando estados e com prazo curto
cenario_b = {
    'price': 1200.00,
    'freight_value': 150.00,
    'product_weight_g': 30000.0,    # 30kg (ex: uma geladeira ou móvel)
    'product_volume_cm3': 150000.0, # Muito grande
    'is_interstate': 1,             # Estados diferentes
    'estimated_days_to_deliver': 5, # Apenas 5 dias de prazo
    'purchase_day_of_week': 4       # Sexta-feira (véspera de fim de semana)
}

print("\n--- TESTE: CENÁRIO A (ALTO RISCO) ---")
prever_atraso(cenario_b)
```

Simulação de um Pedido de Baixo Risco

```python
# Pedido leve, barato, no mesmo estado e com prazo folgado
cenario_a = {
    'price': 50.00,
    'freight_value': 15.00,
    'product_weight_g': 500.0,      # 500g
    'product_volume_cm3': 1000.0,   # Pequeno
    'is_interstate': 0,             # Mesmo estado (SP para SP, por exemplo)
    'estimated_days_to_deliver': 15, # Prazo de 15 dias
    'purchase_day_of_week': 1       # Terça-feira
}

print("--- TESTE: CENÁRIO B (BAIXO RISCO) ---")
prever_atraso(cenario_a)
```
